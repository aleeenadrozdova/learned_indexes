```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('benchmark_results.csv')

search_data = data[data['Operation'] == 'search']
range_search_data = data[data['Operation'] == 'range_search']
build_time_data = data[data['Operation'] == 'build_time']
memory_data = data[data['Operation'] == 'memory']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharey=True)

distributions = ['uniform', 'normal', 'zipf', 'lognormal']

for i, dist in enumerate(distributions):
    ax = axs[i // 2, i % 2]
    subset = search_data[search_data['DistributionType'] == dist]
    
    for index_name in subset['Index'].unique():
        index_data = subset[subset['Index'] == index_name]
        
        index_data = index_data.sort_values('DataSize')
        
        ax.plot(index_data['DataSize'], index_data['Time(ns)'], 
                marker='o', label=index_name)
    
    ax.set_title(f'Search Time - {dist}')
    ax.set_xlabel('Data Size')
    ax.set_ylabel('Time (ns)')
    ax.set_xscale('log') 
    ax.set_yscale('log') 
    ax.legend()

plt.tight_layout()
plt.savefig('search_time_comparison.png')
plt.show()
```
